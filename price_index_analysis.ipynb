```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('updated_data.csv')
```

```python
data
```

#### Avg cost per region

```python
region = []
cost = []

for x in data['Location']:
    region.append(x)
    cost.append(np.mean(data[data['Location'] == x]['AvgCost']))
```

```python
plt.xlabel('Region')
plt.ylabel('Average Cost')
plt.title('Average Cost of Restaurants in Each Region')
sns.barplot(x=region, y=cost)
```

#### Region wise restuarant types

```python
d = {}
for x in data['Location']:
    d[x] = dict(data[data['Location'] == x]['cluster_description'].value_counts())
```

```python
for x in d:
    plt.figure(figsize=(5, 5))
    plt.title(x)
    pallete = sns.color_palette("rainbow", len(d[x]))
    plt.pie(d[x].values(), labels=d[x].keys(), autopct='%1.1f%%', colors=pallete)
    plt.show()
```
