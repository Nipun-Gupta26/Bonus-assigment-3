```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = pd.read_csv('zomatoData.csv')
data_copy = data.copy()
```

```python
data
```

```python
data.isna().sum()
```

```python
sclaer = MinMaxScaler()
data['AvgCost'] = sclaer.fit_transform(data[['AvgCost']])
data['Rating'] = sclaer.fit_transform(data[['Rating']])
```

```python
wcss = []
cluster = []

for i in range(2, 15):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data[['Rating', 'AvgCost']])
    wcss.append(kmeans.inertia_)
    cluster.append(i)
```

```python
plt.plot(cluster, wcss)
```

```python
kmeans = KMeans(n_clusters=6, random_state=0).fit(data.loc[:, ['AvgCost', 'Rating']])
```

```python
data['clusters'] = kmeans.labels_
data_copy['clusters'] = kmeans.labels_
```

```python
sns.scatterplot(data=data, x='Rating', y='AvgCost', hue='clusters')
```

```python
d = {0: 'High Rating and Medium Cost', 1: 'High Rating and Low Cost', 2: 'Low Rating and High Cost', 3: 'Low Rating and Low Cost', 4: 'High Rating and High Cost', 5: 'Low Rating and Medium Cost'}
```

```python
data['cluster_description'] = data['clusters'].map(d)
data_copy['cluster_description'] = data_copy['clusters'].map(d)
```

```python
data.fillna('None', inplace=True)
data_copy.fillna('None', inplace=True)
```

```python
data_copy.to_csv('updated_data.csv', index=False)
```
